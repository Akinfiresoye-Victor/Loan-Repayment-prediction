```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
#Using numpy to generate random data so we can plot them each
import numpy as np
x=np.linspace(0, 10, 25)
y= x*x + 2
print(x)
print(y)
print(np.array([x,y]).reshape(25, 2))
```

```python
#it only takes a single command to draw a graph
pylab.plot(x,y, 'r') #r means red basically colours our lines
```

```python
#creating a subgraph
pylab.subplot(1,2,1) #the content in this bracket follows the order(rows, columns, index) it helps in the formatting of how our data will look like
pylab.plot(x,y, 'r--') #the line will be drawn with red dots
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')#this will be drawn with stars on the line 25 stars to be precise to show where the point was taken
```

```python
#creating a subgraph on the same canvas(blank sheet of paper for drawing)
pylab.subplot(1,2,1)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
#Using figures to plot our graph we can easily edit the size and other things of the graph(we have more tools)
fig = plt.figure() #figure is like a frame or canvas so we can actually interact with the canvas using that

#the axes are used with figures to adjust the size and position of our canvas
axes= fig.add_axes([0.5, 0.1, 0.8, 0.8]) #it follows the order([left, right, width, height]) 

axes.plot(x,y, 'r') #then we use the dimensions we have described above to plot the graph
```

```python
#plotting graphs side by side
fig, axes= plt.subplots(nrows= 1, ncols= 2) #can be above each other just make sure the rows is higher than the column
for ax in axes:
    ax.plot(x,y, 'r')#this plots the graph side by side
```

```python
#drawing a picture or graph inside another graph
fig= plt.figure()
axes1= fig.add_axes([0.1, 0.1, 0.8, 0.8])#initialize the bigger canvas
axes2= fig.add_axes([0.2, 0.5, 0.4, 0.3])#initialize the smaller canvas
axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
#once we plot it we will then see the 2 graphs on the bigger canvas the axes can be used to position the graph where we want to 
```

```python
fig= plt.figure(figsize=(16,9), dpi=100)#dpi increases the quality, the higher the quality the more space it takes and the harder to process
fig.add_subplot()
plt.plot(x, y, 'r')
```

```python
fig, axes= plt.subplots(figsize=(12,9), dpi=100)
axes.set_xlabel('X')#labels for the x axis
axes.set_ylabel('Y')#label for the y axis
axes.set_title('Im a pro')#title of our graph
axes.plot(x,y, 'r')
```

```python
#legend are used as a graph map so we know what each line of our graph entails or holds
ax.legend(['label1', 'label2']) #initializing how many labels we want to put in our graph since we only have 2 then its 2 we put
fig, axes= plt.subplots(figsize=(16,9), dpi=100)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('My first graph')
axes.plot(x, x**2)
axes.plot(x, x**3) #two graph will be plotted but with different colours
#this then will tell the legend what to display the loc=2 actually describes the position we want the legend to be on the graph
axes.legend(["y= x**2", "y= x**3"], loc=2)
```

```python
#in Matplotlib we can actually set othet properties like transperency and line colors and much more
fig, axes= plt.subplots(dpi=150)
axes.plot(x, x**2, color= 'indigo', alpha=0.5) #aplha is what sets the transparency-> range of 0 & 1
axes.plot(x, x+2, color='purple', alpha=0.5)
axes.plot(x, x+3, color='violet')
```

```python
#the dpi is just about the quality of the graph/image we want to draw
fig, ax= plt.subplots(dpi=100)

ax.plot(x, x+1, linewidth=0.25)#we then plot the graph with the linewidth specifying how thick the line is 
ax.plot(x, x+2, linewidth=1)
ax.plot(x, x+3, linewidth=1.5)
ax.plot(x, x+4, linewidth=12)
```

```python
fig, ax= plt.subplots(dpi=100)

#line style just specifies how we want the line to look like on our graph NOTE not everything can work
ax.plot(x, x+1, lw=2, linestyle='-')
ax.plot(x, x+2, lw=2, linestyle='-.')
ax.plot(x, x+3, lw=2, linestyle=':')
line, = ax.plot(x, x+4, color='black', lw=1.50)#lw is the same as line width
line.set_dashes([5, 10, 15, 3])

line, = ax.plot(x, x+6, color='black', lw=1.50)
line.set_dashes([5, 10, 15, 10])

line, = ax.plot(x, x+5, color='black', lw=1.50)
#this set_dashes is just saying that the first line should be 5m long then it will take 10m space then the next line wil be 4m long and so on
line.set_dashes([5, 10, 4, 10])
```

```python
fig, ax= plt.subplots(dpi=100)

#marker helps specify the points where our data is taken
ax.plot(x, x+1, lw=2, marker='.', markersize=5)#the marker size controls the size of our marker
ax.plot(x, x+2, lw=2,marker='s', markerfacecolor='blue')#this just sets the color of the face of our marker 
ax.plot(x, x+3, lw=2)
```

```python
'''Set the canvas grid and axis range'''
#plt.add_subplots(1,2,1) just makes one sup graph with and is positioned using (row,column,position)
#plt.subplots(1,2) makes us able to plot more than one graph at once it is in the format of (row,column)1 by 2 means we will have 2 graphs on 1 row
fig, axes= plt.subplots(1,2, figsize=(10, 5))#figsize is specifying the size of our canvas(space for drawing)
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True) #this just makes grid on our graph

axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].set_ylim([0, 60])#sets limit for our graph so we can limit where the graph will reach
axes[1].set_xlim([2, 5])
axes[1].grid(True)
```

Other 2D Graphics

```python
n= np.array([0,1,2,3,4,5])
print(n)
fig, axes= plt.subplots(1, 4, figsize=(16, 5))
axes[0].set_title('Scatter Plot')#Basic Scatter plot
axes[0].scatter(x, x + 0.25*np.random.random(len(x)))

axes[1].set_title("Step Plot")#step plot
axes[1].step(n, n**2, lw=2)

axes[2].set_title("Bar Plot")#bar plot
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)

axes[3].set_title("Fill_Between Graph") #verry odd plot to me
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
'''Radar chart'''
fig= plt.figure(figsize=(6,6))
ax= fig.add_axes([0.0, 0.0, 0.6, 0.6], polar=True)#polar=true is what makes our chart in a circular form

#this creates an array where the data stats at 0 and ends at 2pi and we wants 100 points within the array so we will have 100 points
t=np.linspace(0, 2 * np.pi, 100)#the np.pi holds the constant of the normal pi and its used here when were working with circle
ax.plot(t,t, color='blue', lw=3)
```

```python
'''Draw a histogram
Histogram basically works on range on the x and y axis of the graph
'''


n= np.random.randn(100000)
fig, axes= plt.subplots(1, 2, figsize=(12,4))

axes[0].set_title("Default Histogram")
axes[0].hist(n)#used for drawing a basic histogram NOTE the bins will be generated automatically if not specified

axes[1].set_title("Cumulative detailed histogram")
#cumulative histogram basically adds up our data just like cumulative graph and stuff like that
axes[1].hist(n, cumulative=True, bins=50)#the bins is telling us how many bar we want in our histogram
```

Contour Image

```python
import matplotlib.cm as cm
```

```python
delta= 0.025
#we want to create and array of data btw -3 and 3 and we want them to have an interval of 0.025(delta)
x= np.arange(-3.0, 3.0, delta)
y=np.arange(-2.0, 2.0, delta)

X, Y= np.meshgrid(x,y)#combines two 1 dimensional array into a 2d array
Z1=np.exp(-X **2 - Y **2)#this is just specifying where our data in the contour will be e^-(x^2 + Y^2) basic 
Z2=np.exp(-(X - 1)**2 -(Y - 1)**2)#just look note abi chatgpt this one hard
Z= (Z1 -Z2) * 2
```

```python
fig, ax = plt.subplots()
CS= ax.contour(X, Y, Z)#if we used contorf we will get the constrast, and contour takes 3 values to point the x,y the coordinate and z being the height
ax.clabel(CS, inline=1, fontsize=10)#inline is telling the contour not to cross the labels
ax.set_title('Contour Map')
```

```python
#Draw 3D surface image
from mpl_toolkits.mplot3d.axes3d import Axes3D#this is used for the 3d

fig= plt.figure(figsize=(14,6))

ax=fig.add_subplot(1, 2, 1, projection='3d')
#the stride(r-row, c-column) is just specifying how tight we want the little squares to be while the line width is showing the thickness of the line of the squares
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
```

Write a python program to create a pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt 

languages= 'Python', 'Java', 'PHP', 'Javascript', 'C#', 'C++' #piechart labels
popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7] #piechart values to follow
colors=['blue', 'yellow', 'purple', 'orange', '#9467bd', '#8c564b'] #the colors of each value
```

```python
explode= (0.3, 0, 0, 0, 0, 0) #this helps seperate or makes a particular slice come out so we can emphasize on it

#labels take our piechat label, autopct prints the value on our slices startangle tells the chart the angle it should start, shadow makes it look cool
plt.pie(popularity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
```
