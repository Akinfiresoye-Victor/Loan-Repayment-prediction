# Decision Tree
Its a tree shaped diagrams where each branch holds possible decisions to be made, it can be used for the classification and the regression
## Terms in Decision tree 
1. Entropy: This is the measure of randomness or unpredictability in the dataset
2. Information gain: Its the measure of decrease in Entropy after the dataset is split (E1-E2)
3. Leaf Node: This carries the classification or the decision it's usualy has 2 or more branches
4. Root Node: THe top most deciion node is known as the root node its where all the data gathers before we split them

## How a decision Tree Work
Problem statement: To classify the different types of animals based on thier features with a decision tree
* Now based on this problem our dataset would be messy due to different types of animal in our dataset  so meaning we have an high entropy, so our training dataset would use color, height and label of the animals, we then frame the conditions that split the data in such a way that the information gain is the highest
* python can solve the logic behind finding the gain because we have to use each conditions and calculate each gain we get for each of this condition we then select the condition which gave us the highest gain as our first split
* So once we use that condition the ones that were true to that condition would go to the true side while the others would go with the false causing our entrropy to reduce considerably after splitting which is good, but we still need to continue splitting each branches until our entropy is equal to 0, this would be when each branc =h actually holds one value
* So now the tree can predict all the classes of animals in the dataset with 100% accuracy(entropy=0)


## Loan Repayment Prediction
Here we want to findout how many customers are going to return the loan they took from the bank or not
Problem Statement: To predict if a customer will repay loan ammount or not

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
#our machine learning training model
from sklearn.tree import DecisionTreeClassifier
#used to evaluate how accurate oue model prediction was
from sklearn.metrics import accuracy_score
#used for the visualization
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
#loading our dataset
balance_data= pd.read_csv('Decision_Tree_ Dataset.csv', sep= ',', header=0)
```

```python
pd.DataFrame(balance_data)
```

```python
print('Dataset length', len(balance_data))
balance_data.shape
```

```python
balance_data.head()
```

```python
#seperating our target variable
X= balance_data.values[:, 0:5] #getting all the numerical value of our dataset
Y= balance_data.values[:, 5] #contains our result which is the yes or no
X
```

```python
#Splitting our dataset into train and set
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.5, random_state= 100)
```

```python
# Creating function to perform training with entropy
clf_entropy= DecisionTreeClassifier(criterion= "entropy", random_state=100, max_depth=3, min_samples_leaf=5)
#does the actual training based on the function we created for training it
clf_entropy.fit(X_train, Y_train)
```

```python
# Testing our predictions
y_pred_en= clf_entropy.predict(X_test)
y_pred_en
```

```python
#Checking Accuracy
print("Accuracy is", accuracy_score(Y_test, y_pred_en)*100)
```

```python
plt.figure(figsize=(12,8))   # set the size of the plot
tree.plot_tree(clf_entropy, 
               filled=True,   
               feature_names=None,
               class_names=True)
plt.show()
```

```python
#TEsting our Predictions on a new customers data
# Testing our predictions
new_data= np.array([30, 89, 300, 200, 120])
new_data= new_data.reshape(1, -1)
y_pred_en= clf_entropy.predict(new_data)
y_pred_en
```
