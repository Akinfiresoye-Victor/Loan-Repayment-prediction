# Random Forest (One of the most powerfull tool in ML)
Random Forest or Random Decision Forest is a method that operates by constructing multiple decision Trees during training phase the decision of majority of the trees is chosen by the random forest as the final decision 
It has higher accuracy and takes lesser time to train, it also provides better object detection in complicated environments

2. Unsupervised learning: Here it has to look at the data and divide it based on its own algorithm withot any training
3. Reinforcement Learning: Here we get a + or - if we get the answer correct

## Advantages of Random Forest
* No Overfitting: use of multiple trees reduces overfitting and takes less time to train
* High Accuracy Runs efficiently with large set of data and still produce very accurate predictions
* Estimates missing data: Can maintain accuracy when a large proportion of data is missing
## Random Forest and decision tree 
In decision tree each beanch represents a possible decision or occurence
### How does random forest work
* So if we have 3 decision trees that splitted fruits and classified them NOTE each of this trees are different
* Tree 1 classifies its own fruit using the diameter(1st split) and color(2nd Split)
* Tree 2 classifies its own using color and shape
* Tree 3 classifies thier own using diameter and if the would grow in summer
* now we have a fruit which has the diameter, shape and if it grows in summer but not the color which is our missing data
* so this fruit is classified using all the decision tree while for the missing color random forest makes sure the fruit falls in the true category if a color is used to classify
* after each of the tree is done classifying random forest then picks the result of the fruit based on maybe if 2 trees said it was lemon and one said it was apple random forest would pick lemon since random forest uses majority votes

## Iris FLower Analysis
NOTE: A problem statement must come first you cant start without a problem
- Problem Statement: To predict what specie of Iris do these flowers belong to

### Data Preparation
#### Importing libraries and dataset 

```python
#importing our iris dataset
from sklearn.datasets import load_iris
#loading our classifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

#setting our random seed
np.random.seed(0)
```

#### Viewing how our data looks like 

```python
#loading our variable with the iris data
iris= load_iris()
#checking all the things we can perform with our iris dataset
print(dir(iris))
#Creating a dataframe with four feature variable
df= pd.DataFrame(iris.data, columns=iris.feature_names)
#viewing the top 5 rows so we see how our data looks like
df.head()
```

#### Getting Our Target for the predictions

```python
#adding a new column so we know what we are actually predicting for (Getting our target)
df['species']= pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

#### Creating our test and train data

```python
#generating a random numbers from 0 to 1 and assigning it to eachof our data in boolean format based on if the random numbers are less than .75
df['is_train']= np.random.uniform(0,1, len(df)) <= .75 #75% data for training 25% data for testing

df.head()
```

```python
#creating dataframes with test rows and training rows
train= df[df['is_train']== True]
test= df[df['is_train']== False]

#showing the number of observation for both train and test dataframes 
print(f'Number of observation in the training data: {len(train)}')
print(f'Number of observation in the testing data: {len(test)}')
```

#### Converting the strings in specie to readable data

```python
#Checking the data we want to work with we dont want to carry the target along just the features of the dataset
features= df.columns[:4]
features
```

```python
y=pd.factorize(train['species'])[0]
y
```

###  Prediction from already prepared data
#### Creating our random forest classifier

```python
#Creating a random forest classifier, n_jobs tells our computer how many cpu cores to use if our data is larger we use more cores
clf= RandomForestClassifier(n_jobs=2, random_state=0)
#training the classifier
clf.fit(train[features], y)
```

#### Testing our classifier

```python
clf.predict(test[features])
```

```python
test[features]
```

#### Predicting our predictions based on majority
each of our data has 3 leaf nodes which is displayed below then random forest tree would then pick the best 2

```python
#viewing how the trees were mapped in this case each data has 3 nodes and as forest works majority carries the vote
clf.predict_proba(test[features])[12:20] #approximation is allowed
```

#### Generating the names for each of our predictions
Instead of us seeing the digits we converted it to previously we just want to convert it back so we can see a more readable prediction/results

```python
#getting all the flower names of our predictions
preds= iris.target_names[clf.predict(test[features])]
preds[0:25]
```

```python
#looking at the first 5 of our actual species
test['species'].head(25)
```

#### Creating a matrix table so we see how many of our data was predicted accurately

```python
#creating a confusion matrix to see how accurate we are
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```

#### Testing our predictions

```python
#predicting a flower based on values we got from a flower outside
preds= iris.target_names[clf.predict( [[5.0, 3.6, 1.4, 2.0 ]] )]
preds
```
