```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
np.random.seed(10)# this helps fix numpy code for reuseability so we get the same value
```

```python
N=30
x= np.random.rand(N) #generates 30 random numbers btw 0 and 1 
y= np.random.rand(N)
colors= np.random.rand(N)
```

```python
area= (30 * np.random.rand(N))**2 #this will multiply each array by 30 so the range will be 0, 30 not 0,1 
# on the graph and the  **2 will make each of the dots bigger
```

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.4) #the alpha helps regulate how visible we want the plots to be
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')

x= [2,4,6]
y= [12, 14, 16]

x2=[3,3,4]
y2= [7,14,5]
```

```python
plt.bar(x, y, color='r', align= 'center')
plt.bar(x2, y2, color= 'b', align= 'center')

plt.title('info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
