## Calculus
### Gradient Descent 
This is how we teach our models by turning their parameters until they behave the way we want them to behave so now basically we wan to get the mininmum and maximum point of a curve/gradient so we can use the minimum point to minimize the error and the maximum point to maximize the use of our output and both this poin can be done using calculus

```python
cur_x= 5 #the algorithm starts at x= 3 on the curve
rate= 0.1 #learning rate it specifies how big the step it should take in the opposite directionof the gradient after each iteration
precision= 0.5 #tells us when to stop the algorithm it basically tells the algorithm that we are close to the minimum point it decides this if the minimumpoint is less than 0.5
previous_step_size= 1 #This measures how far we moved from a point to another
max_iters= 1000000 #the ammount of iterations the algorithm should take
iters= 0 # Iteration counter(Used to count how many interations we have made)
df= lambda x: 2*(x+5) #Gradient of our function

#the continous loop which runs if the previous step is greater than one and the number of iterations is less than the max iter we have set
while previous_step_size > precision and iters < max_iters:
    prev_x= cur_x #stores the current value of x 
    cur_x= cur_x - rate * df(prev_x) #how the grade descent is calculated
    previous_step_size= abs(cur_x - prev_x) #change in x
    iters= iters + 1 #counting the number of iteration
    print("Iteration", iters, "\nX value is", cur_x, previous_step_size)
print("The local minimum occurs at", cur_x)
```

# Statistics 

### Terminology used
* Population - All measurments whose properties are being observed 
* parameter - A metric used to represent a population characteristics
* sample - a subset of the population studied
* variable - a metric of intrest for each person or object in a population

### Classification of sampling
* Random Sampling - selecting 25 people from a list of 250
* Systematic Sampling- selecting 1 employee from every 50 unique employees in a company of 250 employees
* Stratified Sampling - seolecting 1 emp-loyee from every ba=ranch in the company office

### Types of statictics
1. Descriptive Statictics: This is used to describe the nature for our data e.g mean, median, mode, variance etc.
    ##### Measure of spread
   * Range: Difference btw the highest and the lowest value
   * Interquartile Range: Divides a rank-ordered data set into 4 equal parts
   * variance: Measures how far each number in the set is from the meanand also for every other number in the set
   * Standard deviation: the measure of variation or dispersion of a set of values from the mean
3. Inferential Statictics: Used to predict the outcome of the data 

### Descriptive statictics

```python
import pandas as pd 

df= pd.DataFrame({'Name': ['Jane', 'Micheal', 'Willian', 'Rosy', 'Hana', 'Ferdie', 'Graeme'], 'Salary': [50000, 54000, 50000, 189000, 55000, 40000, 59000]})

print(df.head())
```

```python
# Calculate the mean, median, mode

print('Minimum Value is', df['Salary'].min())
print('Maximum Value is', df['Salary'].max())

print('Mean is', df['Salary'].mean())
print('Median is', df['Salary'].median())
print('Mode is', df['Salary'].mode())

print('Range is', df['Salary'].max() -  df['Salary'].min())
```

```python
df['Salary'].describe()
```

```python
#Histogram of salaries

%matplotlib inline
import matplotlib.pyplot as plt

df= pd.DataFrame({'Name': ['Jane', 'Micheal', 'Willian', 'Rosy', 'Hana', 'Ferdie', 'Graeme'], 'Salary': [50000, 54000, 50000, 189000, 55000, 40000, 59000]})

salary= df['Salary']
salary.plot.hist(title='Salary Distribution', color='grey', bins=25)
plt.axvline(salary.mean(), color= 'violet', linestyle= 'dashed', linewidth=2)
plt.axvline(salary.median(), color= 'red', linestyle= 'dashed', linewidth=2)
plt.show()
```

```python
from itertools import product as prod

A= set([x for x in range(1, 7)])
B= set([x for x in range(1, 7)])
p= list(prod(A, B,))# this will show us all the possible combinations of of sset a and b and put it in a list

#this is to get al the possible outcomes of a dice
print("A is set of all possible outcomes of a dice:", A)
print("B is set of all possible outcomes of a dice:", B)
print("\nProduct of A and B is the all possible combinations of A and B thrown together:\n", p)
```

```python
# Throwing 2 dice

n_dice= 2
dice_faces= {1, 2, 3, 4, 5, 6}

#construct the event space i.e set of all POSSIBLE events
event_space= set(prod(dice_faces, repeat=n_dice)) #here we just created a set but the repeat is used to determined how many of that single set we want to work with
event_space
```

```python

for outcome in event_space:
    print(outcome, end=',')
print()
print(len(event_space))#used to know the number of outcomes
```

```python
# Probability to get the multiple of 3, checking if the sum is a multiple of 3

favourable_outcome=[]
for outcome in event_space:
    x,y= outcome #setting each value in each data in a set to both x and y
    if (x+y)%3 == 0: #so if x+y is diveded by 3 we would add both x and y to the list
        favourable_outcome.append(outcome)
favourable_outcome= set(favourable_outcome)
print(favourable_outcome)

prob= len(favourable_outcome)/len(event_space)
# getting the probability of us getting a multiple of 3 when we sum both values we got on a dice
print(f"The probability of getting a sum which is the multiple of 3 is {prob}")
```

```python
#throwing 5 dices

n_dice= 5
dice_faces= {1, 2, 3, 4, 5, 6}

#construct the event space 
event_space= set(prod(dice_faces, repeat=n_dice))
event_space
```

```python
len(event_space)
```

```python
#checking if the sum is a multiple of 5 and not a multiple of 3

favourable_outcome=[]
for outcome in event_space:
    d1,d2,d3,d4,d5= outcome
    if(d1+d2+d3+d4+d5)%5==0 and (d1+d2+d3+d4+d5)%3 !=0:
        favourable_outcome.append(outcome)
favourable_outcome= set(favourable_outcome)
prob=len(favourable_outcome)/ len(event_space)
print(f"The probability of getting a sum which is the multiple of 5 but not a multiple of 3 is {prob}")
print(len(favourable_outcome))
print(favourable_outcome)
```
