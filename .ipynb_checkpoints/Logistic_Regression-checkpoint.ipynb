# Logistic Regression
This is a model used to solve classes classification problem, it helps us predict in a binary way i.e 1, 0 which is yes or No. classification is solved using logistic regression while regression is solved using linear regression logistic gives us a sigmoid curved line showing us how the range of yes and the range of no because Logistic Regressiion uses probability.

Another niche is that logistic regression is used for discrete values(1, Yes, etc.) while linear regression is used for continous values e.g profit

## Steps in Training ou model to recognize digits in a picture from 0-9

### Importing all libraries and their associated methods

```python
# This loads the digits dataset which contains 8X8 pixel imiages of handwriten digits from 0-9
from sklearn.datasets import load_digits
#the model that splits our data set into train and test set for us to work with
from sklearn.model_selection import train_test_split
#to handle the maths
import numpy as np
#Used for the plotting of graphs of our result
import matplotlib.pyplot as plt
import seaborn as sns
#metrics tells us how accurate our prediction was
from sklearn import metrics
%matplotlib inline
digits= load_digits()
```

```python
print("Image Data Shape", digits.data.shape)
print("Label Data Shape", digits.target.shape)
#the 64 shows us thats its an 8X8 matrix
```

### Looking at the data we are working with




```python
plt.figure(figsize=(20,4))

#looping through the first 5(0->4) images to get since our data contains our index while our target contains the images and labels
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    #looping throuigh each data set and putting them into subplots
    plt.subplot(1,5, index+1)
#np.reshape turns the flat vector into an 8X8 grid which is the original grayscale image, so the imshow displays the image using a grayscale color
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    #this aming convetion must be followed if we want to see a proper title
    plt.title('Training =:%i\n' % label, fontsize=20)
```

### Staging our data based for training based on the data set provided

```python
#random state can take any value but each value will carry different numbers
x_train, x_test, y_train, y_test= train_test_split(digits.data, digits.target, test_size= 0.23, random_state=2)
```

```python
print(x_train.shape) #getting the shapes so we know how our trained data looks like
```

```python
print(y_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_test.shape)
```

### The actual Training of our model using LogisticRegression as our training tool

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr= LogisticRegression()
logisticRegr.fit(x_train, y_train)#.fit is what does the actual training using our training library
```

```python
#predicting a particular value based on our trained model NOTE logisticRegr is now the trained model
print(logisticRegr.predict(x_test[0].reshape(1, -1)))
```

```python
print(logisticRegr.predict(x_test[0:10]))
```

```python
#predicting all our data set of our test set
print(logisticRegr.predict(x_test))
```

```python
score= logisticRegr.score(x_test, y_test)
print(score) #finding how accurate our predictions was
```

### Confusion Matrix
this is a way to visually see how and what our trained model predicted it shows the number of accurate predictions the diagonal must have the highest set of values that makes us know that we are actually very accurate so the sum of the diagonal values divided by the sum of our entire values gives us our accuracy NOTE the entire value is the value of our test data set and not our entire dataset

```python
predictions= logisticRegr.predict(x_test)
cm= metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidth=.5, square=True, cmap= "Blues_r");
plt.ylabel("Actual Label");
plt.xlabel('Predicted Label');
all_sample_title= f'Accuracy Score:{score}'
plt.title(all_sample_title, size= 15);
```
